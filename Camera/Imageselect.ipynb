```python
from posix import listdir
import os, os.path
import numpy as np

list_name = os.listdir(File1)
pnglist=[]
for name in list_name:
  if ".png" in name:
    name=name.replace(".png","")
    name = int(name)
    pnglist.append(name)
print(len(pnglist))
```

```python
def DFL(data):
    if len(data)>41:   
      for i in range(20):
        data = np.delete(data, 1, axis=None)

      for j in range(20):
        data = np.delete(data, len(data)-1, axis=None)
    print(len(data))
    return data
```
